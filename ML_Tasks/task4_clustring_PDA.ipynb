```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris         # For loading the Iris dataset
from sklearn.cluster import KMeans              # For performing K-means clustering
from sklearn.decomposition import PCA           # For reducing dimensionality with PCA


```

```python
# 1. Load the Iris dataset
# The Iris dataset contains 150 samples with 4 features each (sepal length, sepal width, petal length, petal width).
data = load_iris()
X = data.data      # Feature matrix of shape (150, 4)
y = data.target    # True labels (not used for clustering, but useful for comparison)

```

```python
# 2. Perform K-means Clustering
# Since the Iris dataset has 3 species, we set n_clusters=3.
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)  # clusters is an array of cluster labels assigned to each sample

```

```python
# 3. Dimensionality Reduction using PCA
# We reduce the 4-dimensional feature space to 2 dimensions for easy visualization.
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

```

```python
# 4. Visualize the Clusters in 2D Space
plt.figure(figsize=(8, 6))
# Create a scatter plot of the PCA-transformed data points.
# The 'c' parameter colors the points based on the K-means cluster labels.
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=50, edgecolor='k')

# Transform the cluster centers from the original space to the PCA-reduced space.
centers_pca = pca.transform(kmeans.cluster_centers_)
# Plot the cluster centers on the same plot.
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', s=200, alpha=0.75, marker='X')
# Add title and labels to the plot for clarity.
plt.title("K-means Clustering on Iris Dataset (PCA-reduced)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()

```
